```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([1,5,2,9,0])
plt.plot(y,marker='*', color = 'orange', linewidth = '3.5')
plt.grid(color = 'green', linestyle = '--', linewidth = 1.5)
plt.show()
```
